### 清洗数据 (额外去掉了粉丝、苹果两个词)

```python
import os
import pandas as pd
import re

vocab = set()
with open('foodNames.txt', 'r', encoding='utf8')as f: # 导入词典
    for line in f:
        word = line.strip()
        vocab.add(word)

def haveFood(sentence, wordMaxLen=6):
    length = len(sentence)
    candDic = {}
    for i in range(length):
        for j in list(range(i + 1, min(i + wordMaxLen + 1, length + 1)))[::-1]:
            cand = sentence[i:j]
            if cand in vocab:
                return True
    return False
def checkLen(text):
    for each in re.split(r'[;,\s]',text):
        if len(each)>18:
            return True
    return False
arr = []
data = pd.read_csv('data.csv', header=0)
print(len(data))
for idx, row in data.iterrows():
    url = row[0]
    text = row[1]
    if len(text)>32 or bool(re.search(r'[\d？【】《》\:\?<>"]', text)) or checkLen(text):
        continue
    if haveFood(text):
        arr.append([url, text])
out = pd.DataFrame(arr)
out = out.sample(frac=1)
out.to_csv('data1.csv', index=False)
print(len(data), len(out))
out.head(10)
```

### 下载图片

```python
import urllib
import pandas as pd
from PIL import Image

def saveImg(url, name):
    try:
        response = urllib.request.urlopen(url) # 得到访问的网址
    except:
        print('Failed', url, name)
        return False
    filename = name + '.jpg'
    img = Image.open(response)
    if img.mode != "RGB":
        img = img.convert('RGB')
    if img.size[0] > 50 and img.size[1] > 50:
        img.save(filename)
        return True
    else:
        return False
    
data = pd.read_csv('data.csv', header=0)
# data.head()
cnt = 1584
for idx, row in data.iterrows():
    if idx<cnt:
        continue
    if idx>=10000:
        break
    url = row[0]
    text = row[1]
    text = text.replace('/',' ').replace('\\',' ')
    if idx % 1000 == 0:
        print(idx)
    path = f'./img/{int(cnt/1000)}/'
    if not os.path.exists(path):
        os.mkdir(path)
    imgpath = os.path.join(path, text)
    if saveImg(url, imgpath):
        cnt += 1
    
    
```

### 划分训练集

```python
import pandas as pd
def split_dataset(path, frac):
    data_df = pd.read_csv(path, header=0)
    # frac-percent; axis=0-get rows; axis=1-get columns
    train_df = data_df.sample(frac=frac,random_state=0,axis=0)
    validation_df = data_df[~data_df.index.isin(train_df.index)]
    return train_df, validation_df
train_df, validation_df = split_dataset('data.csv', 0.9)
print(len(train_df), len(validation_df))

with open('train.txt','w',encoding='utf8')as f:
    for idx, row in train_df.iterrows():
        url = row[0]
        caption = row[1]
        f.write(f'{caption}###{url}\n')
with open('val.txt','w',encoding='utf8')as f:
    for idx, row in validation_df.iterrows():
        url = row[0]
        caption = row[1]
        f.write(f'{caption}###{url}\n')
```

### open_clip 数据预处理

```python
import requests
import os
import multiprocessing as mp
from io import BytesIO
import numpy as np
import PIL
from PIL import Image
import pickle
import sys

ROOT = "cc_data"

def grab(line):
    """
    Download a single image from the TSV.
    """
    uid, split, line = line
    try:
        caption, url = line.split("###")[:2]
    except:
        print("Parse error")
        return

    if os.path.exists(ROOT+"/%s/%d/%d.jpg"%(split,uid%1000,uid)):
        print("Finished", uid)
        return uid, caption, url

    # Let's not crash if anythign weird happens
    try:
        dat = requests.get(url, timeout=20)
        if dat.status_code != 200:
            print("404 file", url)
            return

        # Try to parse this as an Image file, we'll fail out if not
        im = Image.open(BytesIO(dat.content))
        im.thumbnail((512, 512), PIL.Image.BICUBIC)
        if min(*im.size) < max(*im.size)/3:
            print("Too small", url)
            return
        
        if im.mode != "RGB":
            im = im.convert('RGB')
        im.save(ROOT+"/%s/%d/%d.jpg"%(split,uid%1000,uid))

        # Another try/catch just because sometimes saving and re-loading
        # the image is different than loading it once.
        try:
            o = Image.open(ROOT+"/%s/%d/%d.jpg"%(split,uid%1000,uid))
            o = np.array(o)
#             print("Success", o.shape, uid, url)
            return uid, caption, url
        except Exception as e:
            print("Failed", e, uid, url)
            
    except Exception as e:
        print("Unknown error", e)
        pass



if not os.path.exists(ROOT):
    os.mkdir(ROOT)
    os.mkdir(os.path.join(ROOT,"train"))
    os.mkdir(os.path.join(ROOT,"val"))
    for i in range(1000):
        os.mkdir(os.path.join(ROOT,"train", str(i)))
        os.mkdir(os.path.join(ROOT,"val", str(i)))


p = mp.Pool(100)

for txt in ['val.txt']:
    print("Processing file", txt)
    assert 'val' in txt.lower() or 'train' in txt.lower()
    split = 'val' if 'val' in txt.lower() else 'train'
    results = p.map(grab,
                    [(i,split,x) for i,x in enumerate(open(txt).read().split("\n"))])

    out = open(txt.replace(".txt","_output.csv"),"w")
    out.write("title\tfilepath\n")

    for row in results:
        if row is None: continue
        id, caption, url = row
        fp = os.path.join(ROOT, split, str(id % 1000), str(id) + ".jpg")
        if os.path.exists(fp):
            out.write("%s\t%s\n"%(caption,fp))
        else:
            print("Drop", id)
    out.close()

p.close()
```
